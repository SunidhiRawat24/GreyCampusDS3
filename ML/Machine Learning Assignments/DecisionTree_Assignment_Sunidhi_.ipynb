```python
#importing the required libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#reading the csv file 
data = pd.read_csv("C:/Users/sunid/Downloads/IceCreamData.csv")
data
```

```python
Temp = data["Temperature"].values
Temp
```

```python
Rev = data["Revenue"].values
Rev
```

```python
from sklearn.model_selection import train_test_split
Temp_train,Temp_test,Rev_train,Rev_test = train_test_split(Temp,Rev , test_size = 0.3, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(max_depth = 3)
DTR.fit(Temp_train.reshape(-1,1) , Rev_train.reshape(-1,1))
```

```python
#checking score of training as well as testing
print(DTR.score(Temp_train.reshape(-1,1), Rev_train.reshape(-1,1)))
print(DTR.score(Temp_test.reshape(-1,1),Rev_test.reshape(-1,1)))
```

```python
pred = DTR.predict(Temp_test.reshape(-1,1))
```

```python
#creating data frame 
Df = pd.DataFrame({"Actual Temprature":Temp_test.reshape(-1),"Actual Revenue":Rev_test.reshape(-1),"Predicted Revvenue" : pred.reshape(-1)})
Df
```

```python
plt.scatter(Temp_test , Rev_test , color  = "Red")
plt.scatter(Temp_test, pred , color = "Green")
plt.show #data visiualisation 
```
