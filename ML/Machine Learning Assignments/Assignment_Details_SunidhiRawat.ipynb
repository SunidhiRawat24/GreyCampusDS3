**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
More_Mean = data[data["class"]== ">=50K"]["age"].mean()
More_SD = data[data["class"] == ">=50K"]["age"].std()
print("Mean Of more than 50k ",More_Mean)
print("Standard Deviation more than 50k ",More_SD)
Less_Mean = data[data["class"]== "<=50K"]["age"].mean()
Less_SD = data[data["class"] == "<=50K"]["age"].std()
print("Mean Of more than 50k ",Less_Mean)
print("Standard Deviation more than 50k ",Less_SD)
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
m = ["Married-civ-spouse", "Married-spouse-absent","Married-AF-spouse"]
status= data[data["class"]== ">50K"]["marital-status"].isin(m).value_counts()
if status[True] > status[False]:
    print("Married")
else:
    print("Bachelors")
    
    
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
maxhrs = data['hoursperweek'].max()
total = data[(data['hoursperweek']==maxhrs)]['hoursperweek'].count()
more_50 = data[(data['hoursperweek']==maxhrs) & (data['class']=='>50K')]['hoursperweek'].count()
more_50
per = (more_50/total) *100
per
```
